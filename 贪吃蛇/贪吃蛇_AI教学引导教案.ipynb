# 🐍 贪吃蛇游戏项目教学

## 🧭 导语

引导学生从零开始动手开发一款贪吃蛇游戏，并在过程中逐步建立起“编程解决问题”的思维方式。

我们不仅关注能否做出游戏，更注重让学生思考：
- 游戏中的每一个机制是如何实现的？
- 我该如何把“想法”变成“代码”？
- 如何通过拆解问题、调试、优化，让程序更智能？

项目中最后还引入了一个简易 AI 控制机制，引导学生思考“智能”的基础含义。

## 📦 项目阶段结构

| 阶段 | 内容 | 学习重点 | 思维训练 |
|------|------|-----------|-----------|
| 第1步 | 绘制棋盘与蛇 | 坐标系理解、图形编程 | 游戏构成思维 |
| 第2步 | 控制蛇移动 | 主循环、事件处理 | 时间控制、状态变化 |
| 第3步 | 添加食物逻辑 | 随机生成、碰撞检测 | 条件触发、事件响应 |
| 第4步 | 增加死亡判断 | 边界与自身碰撞 | 边界条件判断 |
| 第5步 | 引入简单AI | 贪吃蛇自动吃食物 | 策略与路径思考 |

## 🔹 第一步：绘制棋盘与初始蛇

### 💡 引导问题：
- 如果你要画一个蛇身体上的格子，怎么描述它的位置？
- 坐标系里 `(x, y)` 是怎么定位的？

### 🎯 实现目标：
- 创建一个窗口（使用 `pygame`）
- 蛇的身体由多个方格组成，可以用列表表示

### 🧠 思维训练：
- 学生要学会将“抽象的蛇”变成“一个个有坐标的格子”
- 鼓励学生尝试画出格子编号，锻炼二维空间思维


## 🔹 第二步：让蛇动起来！

### 💡 引导问题：
- 游戏中的“移动”是什么意思？是画面移动，还是数据移动？
- 蛇的每一帧变化是怎么控制的？

### 🎯 实现目标：
- 添加方向控制（键盘输入）
- 主循环中不断刷新蛇的状态

### 🧠 思维训练：
- 认识“游戏循环”与“帧率”的关系
- 理解列表的插入与删除：蛇头移动，尾巴消失

## 🔹 第三步：吃食物

### 💡 引导问题：
- 如何判断蛇“吃到了”食物？
- 如果吃到，蛇应该发生什么变化？

### 🎯 实现目标：
- 在屏幕上随机生成食物
- 如果蛇头与食物位置重合：增加蛇的长度，并刷新食物

### 🧠 思维训练：
- 强化“判断条件”的应用能力
- 引导学生在屏幕与数据结构之间建立联系


## 🔹 第四步：死亡判断

### 💡 引导问题：
- 蛇可能在哪些情况下“死掉”？
- 怎样让程序检测出“死亡”？

### 🎯 实现目标：
- 撞墙判断：超出边界
- 撞自己判断：头部坐标和身体重复

### 🧠 思维训练：
- 强调“边界条件”的重要性
- 鼓励学生调试：不断试错，找出逻辑错误


## 🤖 第五步：加入一个简单 AI！

我们设计一个非常简单的“自动控制蛇吃食物”的策略。虽然它不聪明，但这是 AI 的开始！

### 💡 引导问题：
- 如果你是蛇，你该怎么“决定往哪边走”？
- 你能否编写一个函数，根据蛇头和食物的位置，返回方向？

### 🎯 实现目标：
- 蛇会自动靠近食物（沿x或y方向移动）
- 不使用键盘控制

### 🧠 思维训练：
- 将“策略”变成“代码逻辑”
- 理解“目标驱动”的行为模型
- 引入“路径选择”的朴素概念


## 🚀 拓展挑战任务

- 加入计分系统（每吃一个食物+1）
- 添加界面按钮，如“重新开始”
- 让AI避开自己身体（难度提升）
- 给蛇添加不同速度等级
